# ¿Qué estaciones se usan más y que estaciones se usan menos? ¿Podemos sacar alguna conclusión de alguno de los resultados?

```python
def num_bicicletas_estacion(datos):
    auxiliar = list(datos)
    l = []
    for i in range(1,max(data_final2['estacion de salida'])+1):
        l.append(auxiliar.count(i))
    return [['mayor estacion', l.index(max(l)), max(l)],['menor estacion',l.index(min(l)), min (l)]]
```

```python
num_bicicletas_estacion(data_final2['estacion de salida'])
```

La estacion con más bicis aparcadas sería la estación 42 con 6891 bicicletas aparcadas

En esta sección pintaremos la ruta de unos cuantos de ellos y tambien pintaremos un mapa de calor de las primeros puntos del recorrido de las bicicletas.

Nuestra intencion era pintarlos todos en el mapa de calor para ver por dónde se podía mover la gente pero por temas de memoria y tiempo de respuesta de Python hemos decidido hacerlo así, y además hemos decidido coger de todos los datos que tenemos, una muestra aleatoria simple de 20000 observaciones.

```python
import numpy as np
```

```python
rutas = data_final2['recorrido'].sample(20000).reset_index(drop = True)
```

Vamos a crear dos funciones:
- Una primera función que devuelve una tupla con la latitud y la longitud de cada uno de los 'trackeos' que ha hecho la bicicleta de las posiciones en ciertos instantes de tiempo
- Una segunda funcion donde dado una lista de geometrías, devuelva una tupla con la latitud y longitud del primer 'trackeo' de cada geometría (que vendría a ser cada bicicleta)

Con la primera pintaremos un mapa de rutas y con la segunda realizaremos un mapa de calor

```python
def puntos_de_una_ruta(una_ruta):
    points = []
    tamaño = una_ruta.get('features')
    for i in range (len(tamaño)):
        long = una_ruta.get('features')[i].get('geometry').get('coordinates')[0]
        lat = una_ruta.get('features')[i].get('geometry').get('coordinates')[1]
        points.append(tuple([lat,long]))
    return points
```

```python
#puntos_de_una_ruta(rutas[0])
```

```python
def primer_punto(lista_rutas):
    points = []
    for i in range(len(lista_rutas)):
        long = lista_rutas[i].get('features')[0].get('geometry').get('coordinates')[0]
        lat = lista_rutas[i].get('features')[0].get('geometry').get('coordinates')[1]
        points.append(tuple([lat,long]))
    return points
```
